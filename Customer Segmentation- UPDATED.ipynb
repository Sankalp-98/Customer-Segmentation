```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Train.csv')
df
```

```python
df.isna().sum()
```

```python
df['Ever_Married'].mode()
```

```python
df['Ever_Married']=df['Ever_Married'].fillna('Yes')
df.isna().any()
```

```python
df['Graduated']=df['Graduated'].fillna('Yes')
df.isna().any()
```

```python
df['Profession'].mode()
```

```python
df['Profession']=df['Profession'].fillna('Artist')
df.isna().any()
```

```python
df['Work_Experience'].median()
```

```python
df['Work_Experience']=df['Work_Experience'].fillna(1.0)
df.isna().any()
```

```python
df['Family_Size'].median()
```

```python
df['Family_Size']=df['Family_Size'].fillna(3.0)
df.isna().any()
```

```python
df['Var_1'].mode()
```

```python
df['Var_1']=df['Var_1'].fillna('Cat_6')
df.isna().any()
```

```python
df1=df.copy()
df1
```

```python
df1['Segmentation'].unique()
```

```python
X=df1.drop(columns=['ID','Segmentation'])
X
```

```python
new_X=pd.get_dummies(X)
new_X
```

```python
y=df1['Segmentation'].map({'A':1,'B':2,'C':3,'D':4})
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s1=StandardScaler()
```

```python
new=s1.fit_transform(new_X)
```

```python
new=pd.DataFrame(new)
new
```

```python
new.columns=new_X.columns
```

```python
new.index=new_X.index
```

```python
new
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(new,y,test_size=0.2)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(multi_class='ovr')
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```
